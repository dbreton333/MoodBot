```python
import cv2
import numpy as np
import copy
from matplotlib import pyplot as plt
import os
```

```python
directory = 'drive/MyDrive/Colab/project_Final/data/all'
for filename in sorted(os.listdir(directory)): 
  try:
    img = cv2.imread('drive/MyDrive/Colab/project_Final/data/all/' +filename)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    _,thresh = cv2.threshold(gray,1,255,cv2.THRESH_BINARY)
    contours,hierarchy = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    lst_contours = []
    for cnt in contours:
      ctr = cv2.boundingRect(cnt)
      lst_contours.append(ctr)
    x,y,w,h = sorted(lst_contours, key=lambda coef: coef[3])[-1]
    crop = img[y:y+h,x:x+w]
    cv2.imwrite('drive/MyDrive/Colab/project_Final/data/test/'+filename,crop)
  except:
    continue
```
