```python
import tensorflow as tf
import tensorflow.keras as K
from tensorflow import keras
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import time
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import VGG16
from keras.applications.resnet50 import ResNet50
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
import matplotlib.pyplot as plt
from keras.layers import MaxPooling2D
```

```python
image_filename = []
gazelist1 = []
gazelist2 = []
image_id = []
directory = 'drive/MyDrive/Colab/project_Final/data/test'
for filename in sorted(os.listdir(directory)):
  try:
    if filename.endswith('.jpg'):
      image_filename.append(filename)
      gaze, id = filename.split('_')
      gazelist1.append(float(gaze.split(',')[0]))
      gazelist2.append(float(gaze.split(',')[1]))
      index, trash = id.split('.')
      image_id.append(index)
  except:
    total = [len(gazelist1),len(gazelist2),len(image_id),len(image_filename)]
    minimum = min(total)
    for i in total:
      if i != minimum:
        del i[-1]
```

```python
datadict = {'image_filename' : image_filename, 'gaze1':gazelist1, 'gaze2':gazelist2, 'id':image_id}
```

```python
df = pd.DataFrame(datadict)
```

```python
df = df.sort_values(by='id')
```

```python
df.shape
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df = df.sample(frac=1)
```

```python
df_train = df.iloc[0:30001]
```

```python
df_train
```

```python
df_test = df.iloc[30001:]
```

```python
df_test
```

```python
df_train = df_train.drop(['id'],axis=1)
```

```python
df_test = df_test.drop(['id'],axis=1)
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   fill_mode = "nearest") 

trainGen = train_datagen.flow_from_dataframe(dataframe=df_train, directory=directory, 
                                              x_col="image_filename", y_col=['gaze1','gaze2'], has_ext=True, 
                                              class_mode="raw", target_size=(448, 448), batch_size=32)

testGen = train_datagen.flow_from_dataframe(dataframe=df_test, directory=directory, 
                                              x_col="image_filename", y_col=['gaze1','gaze2'], has_ext=True, 
                                              class_mode="raw", target_size=(448, 448), batch_size=32)
```

```python
new_shape = K.Input(shape=(448,448,3))
```

```python
res_model = ResNet50(
    include_top=False,
    weights="imagenet",
    input_tensor=new_shape,
)
```

```python
len(res_model.layers)
```

```python
for layer in res_model.layers:
  layer.trainable = False
```

```python
model = K.models.Sequential()
model.add(res_model)
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(K.layers.Flatten())
model.add(K.layers.Dense(2048,activation='relu'))
model.add(K.layers.Dropout(rate=0.5))
model.add(K.layers.Dense(1024,activation='relu'))
model.add(K.layers.Dropout(rate=0.5))
model.add(K.layers.Dense(2, activation='relu'))
```

```python
model.compile(loss='mean_absolute_error',
             optimizer= Adam(learning_rate=0.001),
             metrics=['accuracy'])
```

```python
history = model.fit_generator(trainGen,epochs=5,verbose=1,validation_data=testGen)
model.save('drive/MyDrive/Colab/project_Final/src/MoodModel4.h5')
```
